```python
data = [5, 1,23,5,7 , 6 , 8]
```

```python
class node:
    def __init__(self, val , right=None, left=None):
        self.val = val 
        self.right = right
        self.left = left
```

```python
head = node(data[0])
```

```python
def add_node(head, n):

    if head is None:
        return node(n)
    
    if n >= head.val:
        head.right = add_node(head.right, n)
    else:
        head.left = add_node(head.left, n)
    return head
    
```

```python
for d in data[1:]:
    add_node(head, d)
```

```python
head.left.val
head.right.val
```

```python
def draw(head):   
    def dfs(h):
        if h.right:
            G.add_node(h.right)
            G.add_edge(h, h.right)
            dfs(h.right)
        if h.left:
            G.add_node(h.left)
            G.add_edge(h, h.left)
            dfs(h.left)
    
    G = nx.DiGraph()
    dfs(head)
    return G
    
```

```python
G = draw(head)
nx.drawing.nx_pydot.write_dot(G, '/tmp/net.dot')
(graph,) = pydot.graph_from_dot_file('/tmp/net.dot')
graph.write_png('./somefile.png')
```
